```python
import matplotlib.pyplot as plt
from  numpy import NaN
from pandas import Series
```

```python
def demo_interpolate(method: str):
    
    incomplete_values = Series([1, NaN ,3, NaN , 5, NaN, NaN, NaN, 9], index=[1,2,3,4,5,6,7,8,9])
    interpolated_values = incomplete_values.interpolate(method=method)
    
    fig = plt.figure()
    plt.plot(interpolated_values, 'o--')
    plt.plot(incomplete_values,'ro')
    plt.title(f'{method}')
    fig.text(1.0, 0.25, str(interpolated_values), fontsize=16)
    
    
```

```python
methods = ['linear', 'pad', 'nearest']

for each_method in methods:
    demo_interpolate(each_method)
```
